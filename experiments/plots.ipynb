```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, soundfile as sf
import torch, sys, os
from torch.nn.functional import fold, unfold
from torch import arange, FloatTensor, ones
from torch.fft import rfft, irfft
from IPython.display import display, Audio


# From 448 Lab 1 solutions
def stft( x, dft_size, hop_size, zero_pad, window):
    ux = window[:, None] * unfold( x[None, None, None, :], kernel_size=(1, dft_size), stride=(1, hop_size))[0]
    return rfft(ux, n=dft_size+zero_pad, axis=-2)

def istft( f, dft_size, hop_size, zero_pad, window):
    f = window[:, None] * irfft(f, axis=-2)
    return fold(f[None, ...], output_size=(1, f.shape[-1]*hop_size+dft_size-1), kernel_size=(1, dft_size), stride=(1, hop_size))[0,0]

DATA: str = '/mnt/data2/evanmm3/'
```
