# MoviePy-related things!

```python
import numpy as np
import moviepy
import PIL
```

```python
from scipy.io.wavfile import read
from scipy.signal import stft
from matplotlib import pyplot as plt
sr, s = read('../mel/steps.wav')

image_files = []

f, t, Zxx = stft(s, fs=sr, nperseg=512)
print(Zxx.shape)

for i in range(Zxx.shape[1]):
    plt.plot(Zxx[:, i])
    plt.xticks([])
    plt.yticks([])
    plt.title('Giant Steps')
    plt.savefig(f'./frames/steps_frame{i}.jpg')
    plt.clf()
    image_files.append(f'./frames/steps_frame{i}.jpg')
```

```python
import moviepy.video.io.ImageSequenceClip
fps=60
clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(image_files, fps=fps)
```

```python
# video_with_new_audio = moviepy.set_audio(moviepy.AudioFileClip("steps.wav")) 
clip.write_videofile("test.mp4", fps=60, codec="mpeg4")
```
